```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
```

```python
dt=0.25
F=np.matrix([[0,-0.3,0.7],[0.3,0,0],[-0.7,0,0]])
Phi=linalg.expm(F*dt)
print(Phi)
```

```python
N=40
x=np.array([2.3,1.7,6.4])
x_solution=np.zeros((3, N+1))
x_solution[:,0]=np.asarray(x)
for i in range(N+1):
    x=Phi@x
    x_solution[:,i]=x
t=np.linspace(0,10,num=(N+1))
plt.figure(1,figsize=(12, 9))
plt.plot(t, x_solution[0,:], 'r-',label='x1')
plt.plot(t, x_solution[1,:], 'b-',label='x2')
plt.plot(t, x_solution[2,:], 'g-',label='x3')
plt.legend()
plt.show()
```
